#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
# Method 1

employee_grouped = employee.groupby(['Department']).agg({'Salary': 'mean'}).reset_index()
employee_grouped = employee_grouped.sort_values(by='Salary', ascending=False)
print(employee_grouped)
```

```python
# Method 2

employee.pivot_table(index= ["Department"], values = ["Salary"], aggfunc = ["mean"]).reset_index()
```

### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# Define the custom function to remove max salary for each department
def remove_max(salaries):
    max_salary = salaries.max()
    return salaries[salaries < max_salary].mean()

# Create the pivot table with mean salary after removing the maxima for each department
pivot_table = pd.pivot_table(employee, index=['Department', 'Title'], values='Salary', aggfunc=remove_max)

# Print the pivot table
print(pivot_table)

```
